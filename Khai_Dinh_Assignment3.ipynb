```python
# a. Open the file Nebraska Positive Cases Per Date, read it line by line, and close it.

p_case = open('Nebraska Positive Cases Per Date.txt')
lines = p_case.readlines()
p_case.close()

# b. Calculate the total number of positive cases
# Loop through each lines, add the number in each line together, and assigned the value to total

list_num = []
for i in lines:
    int_num = int(i)
    list_num.append(int_num)
print('The total number of positive cases:', sum(list_num), 'cases')


# c. On average, how many people are tested positive per day? 
# devide the total number of cases with the length of the lines

avg = sum(list_num)/len(list_num)
print('On average,', round(avg) , 'people are tested positive per day')

# d. Find maximum number of new cases in a day
# Change value into integer, Append in list, use max function

m_num = max(list_num)
print('The maximum number of new cases per day is:', m_num)


# Use the max function to find the largest number
# Looking for the index of the max value find the date
# create a dictionary 

date_num = list_num.index(m_num) + 2

calendar = {
    0:'March 8th',
    1:'March 10th',
    2:'March 12th',
    3:'March 13th',
    4:'March 14th',
    5:'March 15th',
    6:'March 16th',
    7:'March 17th',
    8:'March 18th',
    9:'March 19th',
    10:'March 20th',
    11:'March 21th',
    12:'March 22th',
    13:'March 23th',
    14:'March 24th',
    15:'March 25th',
    16:'March 27th',
    17:'March 28th',
    18:'March 29th',
    19:'March 30th',
    20:'March 31st',
    21:'April 1st',
    22:'April 2nd',
    23:'April 3rd',
    24:'April 5th',
    25:'April 6th',
    26:'April 7th',
    27:'April 8th',
}

print(calendar[date_num])


```
